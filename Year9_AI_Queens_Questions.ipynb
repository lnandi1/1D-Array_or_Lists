## **(3) Study the Program below and write a Python Program which displays the relative percentage populations of the G7 Countries. Take the data from the program above.**

```python
import matplotlib.pyplot as plt
# Pie chart
Cities = ['London','Paris','Berlin']
People_In_Millions = [8.9,2.1,3.6]
explode = (0.1,0.1,0.2)

fig1, ax1 = plt.subplots()
ax1.pie(People_In_Millions, explode= explode, labels=Cities, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.title("Population in Cities London, Paris, Berlin")

plt.show()
```
